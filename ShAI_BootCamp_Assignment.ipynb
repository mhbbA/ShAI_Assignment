```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
df.columns
```

```python
num_rows, num_columns = df.shape
print(f'Number of rows: {num_rows}')
print(f'Number of columns: {num_columns}')
```

```python
data_types = df.dtypes
print('\nData types of each column:')
print(data_types)
```

```python
missingValues = df.isnull().sum()
print('\nMissing values in each column:')
print(missingValues)
```

```python
# Calculate basic statistics
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode() 
minimum_salary = df['TotalPay'].min()
maximum_salary = df['TotalPay'].max()
range_salary = maximum_salary - minimum_salary
std_deviation = df['TotalPay'].std()
```

```python
print(f'Mean: {mean_salary}')
print(f'Median: {median_salary}')
print(f'Mode: {mode_salary}')
print(f'Minimum: {minimum_salary}')
print(f'Maximum: {maximum_salary}')
print(f'Range: {range_salary}')
print(f'Standard Deviation: {std_deviation}',"\n")
```

```python
#removing theose columns as all values nan
columns_to_remove = ['Status', 'Notes','Benefits']
df_clean_columns = df.drop(columns=columns_to_remove)
```

```python
df_clean_columns
```

```python
#basebay column has a small amount of nan values so it is better to replace it with the mean values
df_clean_columns['BasePay'].fillna(df_clean_columns['BasePay'].mean(), inplace=True)
```

```python
missingValues = df_clean_columns.isnull().sum()
print('\nMissing values in each column:')
print(missingValues)
```

```python
df_clean_columns
#we handle missing values to avoid inacccurate an baised resultes and to improve the accuracy of the over all model
#which improves the model performance
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_clean_columns['TotalPay'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

```

```python
# Visualize the proportion of employees in different departments using a pie chart
dt=df_clean_columns.head(20)
department_counts = dt['JobTitle'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python
# Calculate 'TotalPay' and 'TotalPayBenefits' based on available columns
df_clean_columns=dt
df_clean_columns['TotalPay'] = df_clean_columns['BasePay'] + df_clean_columns['OvertimePay'] + df_clean_columns['OtherPay']
df_clean_columns['TotalPayBenefits'] = df_clean_columns['TotalPay']

# Group the data by 'JobTitle' and calculate summary statistics
grouped_data = df_clean_columns.groupby('JobTitle').agg({
    'TotalPay': ['mean', 'median', 'std', 'min', 'max'],
    'TotalPayBenefits': 'mean'
}).reset_index()

# Display the summary statistics for each JobTitle
print("Summary Statistics by JobTitle:")
print(grouped_data)

# Compare the average total pay across different JobTitles using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='JobTitle', y=('TotalPay', 'mean'), data=grouped_data, palette='pastel')
plt.title('Average Total Pay Across Different JobTitle')
plt.xlabel ('JobTitle')
plt.xticks(rotation='vertical')
plt.ylabel('Average Total Pay')
plt.show()
```

```python
# Calculate 'TotalPay' and 'TotalPayBenefits' based on available columns
df_clean_columns['TotalPay'] = df_clean_columns['BasePay'] + df_clean_columns['OvertimePay'] + df_clean_columns['OtherPay']
df_clean_columns['TotalPayBenefits'] = df_clean_columns['TotalPay']

# Identify correlation between 'TotalPay' and 'Benefits' columns
correlation = df[['TotalPay', 'OvertimePay']].corr()

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='OvertimePay', y='TotalPay', data=df, color='skyblue')
plt.title('Scatter Plot: TotalPay vs OvertimePay')
plt.xlabel('OvertimePay')
plt.ylabel('Total Pay')
plt.show()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation)
```
